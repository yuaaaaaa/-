```python
import sys
import os
import importlib
import cv2
import tkinter as tk
import tkinter.messagebox
from tkinter import filedialog
import tkinter,datetime
from tkinter import *
import time
############################################################################################
#功能描述
##############################################################################################3
# 打开文件夹
def open_familyNumber():
    path = os.getcwd()+'\\face_dataset'
    os.startfile(path)

# 打开存录视频文件夹
def open_video():
    path = os.getcwd()+'\\video'
    os.startfile(path)
    
# 开启设备
def open_camera():
    import os
    os.system('final.py')
    
# 摔倒检测
def fall():
    import os
    os.system('Falldetection\\Fall_down.py')
################################################################################################
# 界面描述
###################################################################################################
importlib.reload(sys)

def uptime():
    global TimeLabel
    TimeLabel["text"] = datetime.datetime.now().strftime('%Y-%m-%d %H:%M:%S:') + "%d" %(datetime.datetime.now().microsecond // 100000)
    window.after(100,uptime)

window = tk.Tk()
window.geometry('350x600+200+10')
window.title('人脸识别小程序')

fm1 = tk.Frame(window)
fm21 = tk.Frame(window)
fm22 = tk.Frame(window)
fm3 = tk.Frame(window)

# Ltop1=tk.Label(fm1,text="wifi", bg='white',anchor="nw")
# Ltop2=tk.Label(fm1,text="12:00", bg='white')

TimeLabel = tk.Label(fm1,width=100,bg='white',font=("times", 20),text = "%s%d" %(datetime.datetime.now().strftime('%Y-%m-%d %H:%M:%S:'),datetime.datetime.now().microsecond // 100000))
TimeLabel.pack(fill=tkinter.BOTH,padx = 10,pady = 8)
window.after(100,uptime)

B1=tk.Button(fm21, text = "点我启动",command = open_camera,width=15,height=5)
B2=tk.Button(fm21, text = "家庭成员",command = open_familyNumber,width=15,height=5)
B3=tk.Button(fm22, text = "摔倒检测",command = fall,width=15,height=5)
B4=tk.Button(fm22, text = "存录视频",command = open_video,width=15,height=5)
Lbot=tk.Label(fm3,text="")

# Ltop1.pack(side = tk.LEFT)
# Ltop2.pack(side = tk.RIGHT)

B1.pack(side=tk.LEFT)
B2.pack(side=tk.LEFT)
B3.pack(side=tk.LEFT)
B4.pack(side=tk.RIGHT)
Lbot.pack(side = tk.BOTTOM)

fm1.pack(side=tk.TOP)
fm21.pack(side=tk.TOP)
fm22.pack(side=tk.TOP)
fm3.pack(side=tk.TOP)
window.mainloop()
```
